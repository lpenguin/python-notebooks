```python
from metadb.api import ElasticApi
```

```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
api = ElasticApi(es, 
                 samples_index='samples_dev',
                 series_index='series_dev'
                )
```

```python
?api.update_by_query
```

```python
len(affy_platform_to_geo)
```

```python
def update_platform(ae_platform, geo_platform):
    samples = [s['accession'] for s in api.search_samples(fields='accession', platform=ae_platform)]
    api.update_by_query(index='samples_dev',
                        conditions=dict(
                            accession=samples
                        ),
                        update=dict(
                            platform=geo_platform
                        )
                       )
```

```python
for ae_platform, geo_platform in affy_platform_to_geo.items():
    print(ae_platform, geo_platform)
    update_platform(ae_platform, geo_platform)
    
```
