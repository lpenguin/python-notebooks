```python
import pandas as pd
import pymongo
```

```python
db = pymongo.MongoClient().scraper_meta
```

```python
pipeline = [
     {"$group": {"_id": "$platform", "count": {"$sum": 1}}},
]

res = list(db.samples.aggregate(pipeline))
```

```python
res[:10]
```

```python
db.platforms.count()
```

```python
platforms = pd.DataFrame(list(db.platforms.find()))
platforms.head()
```

```python
samples_platform = pd.DataFrame(res).rename(columns={'_id': 'accession'})
samples_platform.head()
```

```python
platforms_stats = (
    platforms
    .merge(samples_platform)
    
)
platforms_stats.head()
```

```python
(
    platforms_stats
    [['accession', 'manufacturer', 'title', 'organism', 'count']]
    .sort('count', ascending=False)
)
```
