```python
with connection.cursor() as c:
    sql = "select * from CT_has_Drug"
    c.execute(sql)
    clinical_trials_has_drug = pd.DataFrame(c.fetchall()).assign(drug_id=lambda d: d.Pubchem_id)
clinical_trials_has_drug.head()
```

```python
drugs = pd.read_csv('../data/clinicaltrials/intermediate/pharmatlas_v2_Drug.csv')
drugs.head()
```

```python
clinical_studies = pd.read_csv('../data/clinicaltrials/intermediate/clinicaltrials/clinical_study.csv')
print(clinical_studies.shape, clinical_studies.columns)
clinical_studies.head()
```

```python
clinical_studies['nct_id'] = clinical_studies['nct_id'].str.strip('\n')
```

```python
interventions = pd.read_csv('../data/clinicaltrials/intermediate/clinicaltrials/interventions.csv')
interventions.head()
```

```python
sponsors = pd.read_csv('../data/clinicaltrials/intermediate/clinicaltrials/sponsors.csv')
sponsors.head()
```

```python
clinical_trials_has_drug.shape
```

```python
ct_drugs = (
            clinical_trials_has_drug
            .drop(['id', 'Pubchem_id'], axis=1)
            .merge(drugs
                   .rename(columns=dict(name='drug_name'))
                   .assign(drug_name=lambda d: d.drug_name.str.replace(';', ','))
                   , how='inner', left_on='drug_id', right_on='pubchem_id')
            [['CT_id', 'pubchem_id', 'drug_name']]
            .groupby('CT_id')
            .agg(dict(
                pubchem_id=lambda ids: ';'.join(map(str, ids)),
                drug_name=';'.join
            ))
            .reset_index()
        )
ct_drugs.head()
```

```python
ct_sponsors = (
    clinical_trials_ids
    .merge(sponsors, on='nct_id')
    [['nct_id', 'agency']]
#     .groupby('nct_id')
#     .agg({'agency': ';'.join})
    .reset_index()
)
ct_sponsors.head()
```

```python
(
    clinical_studies
    .merge(clinical_trials_ids)
    .merge(
        ct_drugs, how='inner', left_on='id', right_on='CT_id'
    )
    .merge(ct_sponsors, on='nct_id')
    
    [['nct_id', 
      'brief_title', 
      'detailed_description', 
      'overall_status', 
#       'source', 
      'start_date', 
#       'is_fda_regulated',
      'phase',
      'completion_date',
      'drug_name',
      'pubchem_id',
      'agency'
     ]]
#     .head()
    .to_csv('../data/clinicaltrials/clinicaltrials-phase-no-na-drugs.agg.csv', index=False)
#     .query('nct_id == "NCT00000125"')
#     .drug_name
#     .shape
)
```
