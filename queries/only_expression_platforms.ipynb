```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
```

```python
res = es.search(
    index='samples_dev',
    body={
        'query': {
            "filtered": {
                 "filter": {
                       "terms": {
                            "assigned.experiment_type": [
                              "Expression profiling by array",
                              "Expression profiling by high throughput sequencing"
                            ]
                          }
                }
            }
        },
        'aggs': {
            "platforms": {
              "terms": {
                "field": "platform",
                "size": 7000
              }
            }
          }
        ,
    'size': 0
    }
)
res
```

```python
pls = [b['key'] for b in res['aggregations']['platforms']['buckets']]
```

```python
import pandas as pd
```

```python
pd.Series(pls).to_csv('../data/tmp.expr.platforms.csv', index=False)
```
