```python
from elasticsearch import Elasticsearch
from metadb.api import ElasticApi
es = Elasticsearch()
api = ElasticApi(es, 'samples_dev', 'series_dev')
```

```python
api.samples_count(
    organism='Homo sapiens',
    annotated=1,
    platform=['GPL6947', 'GPL10558']
)
```

```python
samples = list(api.search_samples(
    organism='Homo sapiens',
    annotated=1,
    platform=['GPL6947', 'GPL10558'],
    fields=['accession', 'series', 'platform']
))
```

```python
samples_df = pd.DataFrame(samples)
```

```python
samples_df = utils.expand_dataframe(samples_df, 'series')
```

```python
_t = (
    samples_df
    [['platform', 'series']]
    .drop_duplicates()
    .dropna()
#     .assign(dataset=lambda d: d['series'].combine(d['platform'], lambda x, y: '{}_{}'.format(x, y)))
#     .shape
#     .head()
)
```

```python
res = []
for _, (series, platform) in _t.iterrows():
    res.append('{}_{}'.format(platform, series))
```

```python
res
```

```python
with open('tmp.ill.datasets', 'w') as f:
    for l in res:
        f.write(l+'\n')
```
