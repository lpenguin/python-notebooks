```python
import json
import pandas as pd
import numpy as np
```

```python
json_file = '/data/rawdata/snapshots/mongo/scraper_test_dev.samples.series.json'
```

```python
def read_dump_file(file_name):
    lines_count = 0
    with open(file_name) as f:

            
        for l in f:
            lines_count += 1
            if lines_count % 10000 == 0:
                print("Processed #{} lines".format(lines_count))
            l = l.rstrip("\n")
            data = json.loads(l)
            for series in data['series']:
                yield (data['accession'], series) 
```

```python
samples_series = pd.DataFrame.from_records(read_dump_file(json_file))
samples_series.head()
```

```python
samples_series.to_pickle('../data/geo-annotation/samples_series.v1.pickle')
```
