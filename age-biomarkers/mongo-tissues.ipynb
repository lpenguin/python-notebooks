```python
import pandas as pd
tissues = pd.read_csv('../data/age-biomarkers/tissues.csv').dropna().sort('count', ascending=False)
```

```python
tissues['name'] = tissues['_id'].str.lower()
tissues = tissues.groupby('name').sum().sort('count', ascending=False).reset_index()
tissues[tissues['name'].str.contains('muscle')]
```
