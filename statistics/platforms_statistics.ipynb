```python
import pymongo as pm
import pandas as pd
```

```python
db = pm.MongoClient().scraper_test_dev
```

```python
pipeline = [
    {'$group': {
            '_id': '$platform',
            'count': {'$sum': 1}
        }}
]
```

```python
_c = db.samples.aggregate(pipeline)
samples_per_platform = pd.DataFrame(list(_c))
```

```python
samples_per_platform
```

```python
(
    samples_per_platform
    .sort('count', ascending=False)
)
```

```python
fields = {
    '_id': 0,
    'accession': 1,
    'manufacturer': 1,
    'title': 1
}

_c = db.platforms.find({'organism': 'Homo sapiens'}, fields)
platforms = pd.DataFrame(list(_c))
platforms.head()
```

```python
(
    samples_per_platform
    .merge(platforms, left_on='_id', right_on='accession', how='inner')
    .drop_duplicates(subset=['accession', '_id'])
    .sort('count', ascending=False)
)
```
